### Data visualizations and Insights

```python
# Check distribution of death
%matplotlib inline
sns.distplot(df2["hospitalized"] )
```

#### Statewise details of patients who were tested positive, hospitalized, onVent support, in ICU, total death

```python
#sum numbers by State

state_agg = df2.groupby('state')['positive', 'hospitalized', 'onVentilatorCurrently', 'inIcuCurrently', 'death'].sum()
state_agg.sort_values('death', ascending=False)
```

#### Insight:  
Statewise, patient who are Covid positive and hospitalized, on Ventilator support and IN ICU and finally number of dealth in each state gives good clarity of which are the states that are highly impacted. 

Newyork, california and New Jersey are highly infected with many death. 

#### Statewise details of people who were tested, hospitalized, Discharged, recovered, total death

```python
#sum numbers by State

state_agg_re = df2.groupby('state')['totalTestResults', 'hospitalized', 'hospitalizedDischarged', 'recovered', 'death'].sum()
state_agg_re.sort_values('recovered', ascending=False)
```

#### Insights:

Texas and Ohio has better recovery in terms of numbers. 

```python
#sum numbers by State

state_agg_inc = df2.groupby('state')['positiveIncrease', 'negativeIncrease', 'deathIncrease', 'hospitalizedIncrease'].sum()
state_agg_inc.sort_values('deathIncrease', ascending=False)
```

#### Insight:
Califorina - has highest in increasing death

### Ploting trend of positive, recovered and hopistalized

```python
#Plotting distribution for Positive count, recovered and hospitalized
plt.figure(6)

plt.subplot(131)
df2.groupby('state')['positive'].sum().plot.bar(figsize = (20,12), title = 'state wise postive count')

plt.subplot(132)
df2.groupby('state')['recovered'].sum().plot.bar(figsize = (20,12), title = 'state wise recovered count')

plt.subplot(133)
df2.groupby('state')['hospitalized'].sum().plot.bar(figsize = (20,12), title = 'state wise hospitalized count')
```

### Ploting trend of On ventilator currently, In ICU currently  and death

```python

plt.subplot(231)
df2.groupby('state')['onVentilatorCurrently'].sum().plot.bar(figsize = (20,10), title = 'state wise Ventilator count')

plt.subplot(232)
df2.groupby('state')['inIcuCurrently'].sum().plot.bar(figsize = (20,10), title = 'state wise ICU count')

plt.subplot(233)
df2.groupby('state')['death'].sum().plot.bar(figsize = (20,10), title = 'state wise dealth count')
```
